```python
import numpy as np
import pandas as pd

from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
from tensorflow.keras.utils import to_categorical
```

```python
from google.colab import drive
drive.mount("/content/drive")
```

```python
data = pd.read_excel("/content/drive/MyDrive/Colab Notebooks/Final Project/2024.2025 On-Campus Housing Lottery_Data.xlsx")
data
```

```python
data = data.drop(["Email","Name","Are you interested in staying in Ashesi's halls of residence next academic year?",
         "Full Name (First, Middle, Last Names)","Student ID Number","Phone Number","Are you on any form of financial aid?",
         "What's your financial aid status?","Do you have any special accessibility needs?","What is the specific need?",
         "Please state any special accomodations you will need.","How would your friends describe you in one word?"],axis=1)
data
```

```python
data = data.dropna()
data
```

```python
categorical_cols = ["Gender","Year Group","How would you describe your personality?",
                    "What time do you typically go to bed on weeknights?",
                    "Where do you prefer to do homework/study?","What do you enjoy doing for fun?",
                    "Do you enjoy having friends over in your room?"]
```

```python
#Building the transformer
col_trans = ColumnTransformer(transformers=[('labels',OneHotEncoder(),categorical_cols)])
```

```python
encoded_data = col_trans.fit_transform(data)
encoded_data
```

```python
encoded_labels = col_trans.named_transformers_['labels'].get_feature_names_out(categorical_cols)

final_columns = list(encoded_labels)
final_columns
```

```python
encoded_data.shape
```

```python
encoded_df = pd.DataFrame(encoded_data, columns=final_columns)
encoded_df
```
