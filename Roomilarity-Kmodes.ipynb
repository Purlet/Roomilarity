```python
data = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/Final Project/Roomilarity_dataCAT.csv")
data
```

```python
#since I want the gender to weigh more I'm duplicating the gender category
data['Gender2']=data['Gender']
data['Gender3']=data['Gender']
data
```

```python
centroids = 10
```

```python
kmodes = KModes(n_clusters = centroids, max_iter = 300, random_state=42,verbose = 1)
clusters = kmodes.fit_predict(data)
```

```python
data['cluster'] = clusters
data[data['cluster']==0]
```
