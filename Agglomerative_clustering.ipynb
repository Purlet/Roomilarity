```python
column_train = train_set.columns

print("These are all the features present in the survey: \n", column_train)
```

```python
#Deleting the unneeded columns
columns_to_drop = [
    'ID',
    'Email',
    'Name',
    'Full Name (First, Middle, Last Names)',
    'Phone Number',
    "What's your financial aid status?",
    'Student ID Number',
    "Are you on any form of financial aid?",
    "Do you have any special accessibility needs?"
]

train_set.drop(columns=columns_to_drop, inplace=True)

train_set.columns
```

```python
# Replacing NaNs in the 'Specific needs/Special accomodations' column with No
train_set['What is the specific need?'] = train_set['What is the specific need?'].fillna('No')
train_set['Please state any special accomodations you will need.'] = train_set['Please state any special accomodations you will need.'].fillna('No')
col = "Are you interested in staying in Ashesi's halls of residence next academic year?"

# Overwriting the DataFrame with only the rows where the Ashesi's Halls residence equals 'Yes'
train_set = train_set.loc[train_set[col] == 'Yes'].copy()
train_set = train_set.dropna()
```

```python
cols_to_encode = [
    "Are you interested in staying in Ashesi's halls of residence next academic year?",
    "Gender",
    "Year Group",
    "What is the specific need?",
    "Please state any special accomodations you will need.",
    "How would you describe your personality?",
    "What time do you typically go to bed on weeknights?",
    "Where do you prefer to do homework/study?",
    "What do you enjoy doing for fun?",
    "Do you enjoy having friends over in your room?",
    "How would your friends describe you in one word?"
]

encoders = {}                      # keeping a reference if you’ll need them later

for col in cols_to_encode:
    le = LabelEncoder()
    train_set.loc[:, col] = le.fit_transform(train_set[col])
    encoders[col] = le            # storing for test‑set transforms or inverse mapping

train_set
```

```python
# Splitting data based on gender
train_male = train_set[train_set['Gender'] == 1]
train_female = train_set[train_set['Gender'] == 0]


# Performing clustering on each subset
clustering_male = AgglomerativeClustering(n_clusters=5, linkage='ward')  # Clustering males
clustering_female = AgglomerativeClustering(n_clusters=5, linkage='ward')  # Clustering females

male_labels = clustering_male.fit_predict(train_male)
female_labels = clustering_female.fit_predict(train_female)

# Attaching labels back to each gender-specific subset
train_male['cluster'] = male_labels
train_female['cluster'] = female_labels

train_male
```
