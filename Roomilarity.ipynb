```python
import pandas as pd
import numpy as np
```

```python
data=pd.read_excel('/content/drive/MyDrive/Colab Notebooks/Final Project/2024.2025 On-Campus Housing Lottery_Data.xlsx')
data
```

```python
data.describe()
```

```python
#Extracting Needed colums for Classification
Needed_columns=['Gender','Year Group','Are you on any form of financial aid?','Do you have any special accessibility needs?','How would you describe your personality?','What time do you typically go to bed on weeknights?','Where do you prefer to do homework/study?','What do you enjoy doing for fun?','Do you enjoy having friends over in your room?','How would your friends describe you in one word?']
Needed_data=data[Needed_columns]
Needed_data=pd.DataFrame(Needed_data)
Needed_data
```

```python
#Dropping Missing Values
Needed_data=Needed_data.dropna()
Needed_data
```

```python
#Converting to numerical values to enable k_means classification
from sklearn.preprocessing import LabelEncoder
lab_Enc=LabelEncoder()
```

```python
# Male=1  Female=0
Needed_data['Gender']=lab_Enc.fit_transform(Needed_data['Gender'])

# C2027= 3   C2026= 2   C2025= 1
Needed_data['Year Group']=lab_Enc.fit_transform(Needed_data['Year Group'])

# Yes=1     No=0
Needed_data['Are you on any form of financial aid?']=lab_Enc.fit_transform(Needed_data['Are you on any form of financial aid?'])

# Yes=1     No=0
Needed_data['Do you have any special accessibility needs?']=lab_Enc.fit_transform(Needed_data['Do you have any special accessibility needs?'])

# Extroverted=0   Introverted=1  Somewhere In Between=2
Needed_data['How would you describe your personality?']=lab_Enc.fit_transform(Needed_data['How would you describe your personality?'])

#  Nope= 0  Not really, but I don't mind=1  Yup, all the time=2  Yup, but not always= 3
Needed_data['Do you enjoy having friends over in your room?']=lab_Enc.fit_transform(Needed_data['Do you enjoy having friends over in your room?'])

# Going for events on/off-campus=0   Hanging out with friends=1  Relaxing/Sleeping=2
Needed_data['What do you enjoy doing for fun?']=lab_Enc.fit_transform(Needed_data['What do you enjoy doing for fun?'])

# Away from the room=0   In the room=1
Needed_data['Where do you prefer to do homework/study?']=lab_Enc.fit_transform(Needed_data['Where do you prefer to do homework/study?'])

# 10PM or earlier=0  After Midnight (12AM)=1 Between 10PM and Midnight (12AM) =1
Needed_data['What time do you typically go to bed on weeknights?']=lab_Enc.fit_transform(Needed_data['What time do you typically go to bed on weeknights?'])

#  Chill=2   Friendly=4 Fun=5 Organized=8 Responsible=10
Needed_data['How would your friends describe you in one word?']=lab_Enc.fit_transform(Needed_data['How would your friends describe you in one word?'])



```

```python
clean_data=Needed_data
clean_data
```

```python
#Using the elbow method to find the optimal k for clustering
from sklearn.cluster import KMeans
import  matplotlib.pyplot as plt
```

```python
features=clean_data[['How would you describe your personality?']]
k_values=range(1,11)
WSS=[]

for i in k_values:
  kmeans=KMeans(n_clusters=i)
  kmeans.fit(features)
  WSS.append(kmeans.inertia_)
```

```python
plt.plot(k_values,WSS,'bo-')
plt.xlabel('Number of clusters(k)')
plt.ylabel('Wihtin Cluster Sum of Squares(WSS)')
plt.title('Elbow Method for Optimal k')
plt.xticks(k_values)
plt.grid(True)
plt.show()



```

```python
#From the graph, the optimal k value is 3
```

```python
kmeans=KMeans(n_clusters=3)
kmeans.fit(features)
```

```python
clean_data['kmeans_3']=kmeans.labels_
clean_data
```

```python
#Visualizing created Clusters
plt.scatter(clean_data['How would you describe your personality?'],clean_data['kmeans_3'])
plt.xlabel('Personality')
plt.ylabel('Cluster')
plt.title('Personality Clusters')
plt.show()
```
